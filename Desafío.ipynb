#### 1) b)	Calcular la redemption rate SIMPLE: proporci√≥n de clientes que al menos compraron 1 producto de la promo (sin importar la cantidad).

```python
# Calcular clientes redimidos por grupo
redimidos_simple = ventas_con_promos.groupby('MEASUREMENT_GROUP')['POC_ID'].nunique()

# Total de clientes con promoci√≥n por grupo
total_clientes_promocion = promociones.groupby('MEASUREMENT_GROUP')['POC_ID'].nunique()

# Redemption rate SIMPLE
redemption_rate_simple = redimidos_simple / total_clientes_promocion
print("Redemption Rate SIMPLE:\n", redemption_rate_simple)
```

1) c) Calcular la redemption rate REAL: % clientes que compraron una cantidad de unidades mayor o igual a la asignada por el algoritmo.


```python
# Filtrar ventas que cumplen con la cantidad m√≠nima
ventas_cumplen_min_qty = ventas_con_promos[
    ventas_con_promos['UNIDADES'] >= ventas_con_promos['MIN_QTY']
]

# Calcular clientes redimidos que cumplen con la cantidad m√≠nima
redimidos_real = ventas_cumplen_min_qty.groupby('MEASUREMENT_GROUP')['POC_ID'].nunique()

# Redemption rate REAL
redemption_rate_real = redimidos_real / total_clientes_promocion
print("Redemption Rate REAL:\n", redemption_rate_real)
```

1) d) Comparar el volumen total y el volumen promedio. Previamente eliminar outliers.


```python
# Eliminar outliers basados en volumen (usar IQR para detectar y eliminar outliers)
Q1 = ventas_con_promos['TOTAL_VOLUME'].quantile(0.25)
Q3 = ventas_con_promos['TOTAL_VOLUME'].quantile(0.75)
IQR = Q3 - Q1
ventas_sin_outliers = ventas_con_promos[
    (ventas_con_promos['TOTAL_VOLUME'] >= (Q1 - 1.5 * IQR)) &
    (ventas_con_promos['TOTAL_VOLUME'] <= (Q3 + 1.5 * IQR))
]

# Calcular volumen total y promedio por grupo
volumen_total = ventas_sin_outliers.groupby('MEASUREMENT_GROUP')['TOTAL_VOLUME'].sum()
volumen_promedio = ventas_sin_outliers.groupby('MEASUREMENT_GROUP')['TOTAL_VOLUME'].mean()

print("Volumen Total:\n", volumen_total)
print("Volumen Promedio:\n", volumen_promedio)
```

An√°lisis de Comportamiento Hist√≥rico
3.1 Alcanzabilidad de Cantidades M√≠nimas
Analizar el historial de compras para evaluar si las cantidades m√≠nimas exigidas por el algoritmo son alcanzables.

```python
# Calcular historial de compras por cliente y SKU
historial_compras = ventas.groupby(['POC_ID', 'SKU'])['UNIDADES'].mean()

# Comparar con las cantidades m√≠nimas del algoritmo
alcanzabilidad = promociones.merge(historial_compras, on=['POC_ID', 'SKU'], how='left')
alcanzabilidad['cumple_min_qty'] = alcanzabilidad['UNIDADES'] >= alcanzabilidad['MIN_QTY']

# Proporci√≥n de cumplimiento de cantidades m√≠nimas por grupo
cumplimiento_por_grupo = alcanzabilidad.groupby('MEASUREMENT_GROUP')['cumple_min_qty'].mean()

print("Proporci√≥n de Cumplimiento de Cantidades M√≠nimas:\n", cumplimiento_por_grupo)
```

# Clustering de Clientes
Clusterizar los clientes seg√∫n variables relevantes como volumen total, cantidad de SKU‚Äôs comprados, y frecuencia de compra.

```python
from sklearn.cluster import KMeans
import numpy as np

# Variables para clustering
clientes = ventas.groupby('POC_ID').agg({
    'TOTAL_VOLUME': 'sum',
    'SKU': 'nunique',
    'ORDER_DATE': 'count'
}).rename(columns={'SKU': 'NUM_SKUS', 'ORDER_DATE': 'FREQUENCY'})

# Normalizaci√≥n de datos
clientes_normalized = (clientes - clientes.mean()) / clientes.std()

# Aplicar KMeans
kmeans = KMeans(n_clusters=3, random_state=42).fit(clientes_normalized)
clientes['CLUSTER'] = kmeans.labels_

# Analizar performance por cluster
ventas_con_clusters = ventas_con_promos.merge(clientes[['CLUSTER']], on='POC_ID')
redemption_rate_cluster = ventas_con_clusters.groupby(['MEASUREMENT_GROUP', 'CLUSTER'])['POC_ID'].nunique() / total_clientes_promocion
volumen_cluster = ventas_con_clusters.groupby(['MEASUREMENT_GROUP', 'CLUSTER'])['TOTAL_VOLUME'].sum()

print("Redemption Rate por Cluster:\n", redemption_rate_cluster)
print("Volumen Total por Cluster:\n", volumen_cluster)
```
